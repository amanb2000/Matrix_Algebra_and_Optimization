```julia
using LinearAlgebra
using Plots
using Plotly
using SymPy

plotly(size = (500,500))
```

```julia
# Ellipses, Part 1: Finding Eigen (Spectral) Decomposition:

P = [9 -2; -2 6]
x_c = [-2; 1]

λ, U = eigen(P)
```

```julia
# Showing that the eigen equality holds for each vector.
println("Equality for first vector: ", P*U[1,:] == λ[1]*U[1,:])
println("Equality for second vector: ", P*U[2,:] == λ[2]*U[2,:])
```

```julia
# Forming our Lambda matrix
Λ = zeros(2,2)
Λ[1,1] = λ[1]
Λ[2,2] = λ[2]
Λ
```

```julia
c = 1 # constant that our quadratic is limited by:
```

```julia
x = Sym("x")
ex = sqrt(λ[2]-(λ[2]x^2)/λ[1])
ex_2 = -1*sqrt(λ[2]-(λ[2]x^2)/λ[1])

xs = collect(-10:0.001:10)

ys1 = ex(xs)

Plots.plot(ex, xlims=(-2,2), ylims=(-2,2))        # plot of a symbolic expression
Plots.plot!(ex_2, xlims=(-2,2), ylims=(-2,2))
Plots.title!("x̃ Space")
```

```julia
ys1 = zeros(length(xs))

for i = 1:length(ys1)
    ys1[i] = ex(x=>xs[i])
end

ys2 = zeros(length(xs))

for i = 1:length(ys2)
    ys2[i] = ex_2(x=>xs[i])
end
```

```julia
Plots.plot(xs, ys1, xlims=(-2,2), ylims=(-2,2))
Plots.plot!(xs, ys2, xlims=(-2,2), ylims=(-2,2))
Plots.title!("Still in squiggle X space!")
```

```julia
comb_mat_1 = [transpose(xs); transpose(ys1)]
comb_mat_2 = [transpose(xs); transpose(ys2)]
```

```julia
trans_comb_mat_1 = U*comb_mat_1
trans_comb_mat_2 = U*comb_mat_2
```

```julia
Plots.plot(trans_comb_mat_1[1,:], trans_comb_mat_1[2,:], xlims=(-2,2), ylims=(-2,2))
Plots.plot!(trans_comb_mat_2[1,:], trans_comb_mat_2[2,:], xlims=(-2,2), ylims=(-2,2))
Plots.title!("Transformed using U")
```

```julia
println("Next: just translate back using x_0")
```

```julia
size(comb_mat_1,2)

for i = 1:size(comb_mat_1,2)
    trans_comb_mat_1[:,i] += x_c
    trans_comb_mat_2[:,i] += x_c
end
```

```julia
Plots.plot(trans_comb_mat_1[1,:], trans_comb_mat_1[2,:], xlims=(-5,5), ylims=(-5,5))
Plots.plot!(trans_comb_mat_2[1,:], trans_comb_mat_2[2,:], xlims=(-5,5), ylims=(-5,5))
Plots.title!("Final Elipse")
```
