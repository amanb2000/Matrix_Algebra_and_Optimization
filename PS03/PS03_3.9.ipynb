```julia
using LinearAlgebra
using Plots
using Plotly
using SymPy

plotly()
```

```julia
# Ellipses, Part 1: Finding Eigen (Spectral) Decomposition:

P = [1.5 -0.5; -0.5 1.5]
x_c = [0; 0]

λ, U = eigen(P)
```

```julia
# Showing that the eigen equality holds for each vector.
println("Equality for first vector: ", P*U[1,:] == λ[1]*U[1,:])
println("Equality for second vector: ", P*U[2,:] == λ[2]*U[2,:])
```

```julia
# Forming our Lambda matrix
Λ = zeros(2,2)
Λ[1,1] = λ[1]
Λ[2,2] = λ[2]
Λ
```

```julia
c = 1 # constant that our quadratic is limited by:
```

```julia
x = Sym("x")
ex = sqrt(λ[2]-(λ[2]x^2)/λ[1])
ex_2 = -1*sqrt(λ[2]-(λ[2]x^2)/λ[1])

xs = collect(-1:0.05:1)

ys1 = ex(xs)

Plots.plot(ex)        # plot of a symbolic expression
Plots.plot!(ex_2)
Plots.title!("x̃ Space")
```

```julia
ys1 = zeros(length(xs))

for i = 1:length(ys1)
    ys1[i] = ex(x=>xs[i])
end

ys2 = zeros(length(xs))

for i = 1:length(ys2)
    ys2[i] = ex_2(x=>xs[i])
end
```

```julia
Plots.plot(xs, ys1)
Plots.plot!(xs, ys2)
Plots.title!("Still in squiggle X space!")
```

```julia
comb_mat_1 = [transpose(xs); transpose(ys1)]
comb_mat_2 = [transpose(xs); transpose(ys2)]
```

```julia
trans_comb_mat_1 = U*comb_mat_1
trans_comb_mat_2 = U*comb_mat_2
```

```julia
Plots.plot(trans_comb_mat_1[1,:], trans_comb_mat_1[2,:])
Plots.plot!(trans_comb_mat_2[1,:], trans_comb_mat_2[2,:])
Plots.title!("Transformed using U")
```
