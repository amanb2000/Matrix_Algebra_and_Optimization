```julia
### IMPORT BOX ###

using LinearAlgebra
using Plots
using JuMP
import GLPK
```

```julia

#######################
# GENERATING MATRIX M #
#######################
ary_size = 10 # Resolution of our control system's control.
M = zeros(ary_size,ary_size)

row = zeros(1,ary_size)
row[1] = 0.5

for i = 1:ary_size
    M[i,:] = row
    for j = 1:ary_size-1
        row[ary_size-j+1] = row[ary_size-j]
    end
    row[1] += 1
end

M
```

```julia
# Helper matrix to determine speed vector given control force vector #

B = zeros(ary_size,ary_size)

row = zeros(1,ary_size)
row[1] = 1

for i = 1:ary_size
    B[i,:] = row
    for j = 1:ary_size-1
        row[ary_size-j+1] = row[ary_size-j]
    end
end

B
```

```julia
# Creating constraint matrix and running least squares to find smallest l2 norm solution:

A = zeros(2,ary_size)
A[1,:] = M[10,:]
A[2,:] = ones(1,ary_size)

y = [1; 0]

A
```

## $l_1$ Solution

```julia
model = Model(GLPK.Optimizer)

clctr = ones(10)

@variable(model, p[1:10])
@variable(model, t[1:10]) # Helper variable

# @objective(model, Min, t[1]+t[2]+t[3]+t[4]+t[5]+t[6]+t[7]+t[8]+t[9]+t[10])
@objective(model, Min, sum(t))

@constraint(model, sum(p) == 0)

@constraint(model, 9.5p[1]+8.5p[2]+7.5p[3]+6.5p[4]+5.5p[5]+4.5p[6]+3.5p[7]+2.5p[8]+1.5p[9]+0.5p[10] == 1)

@constraint(model, p .<= t)
@constraint(model, p .>= -t)

@constraint(model, t .>= 0)


print(model)
```

```julia
optimize!(model)
println("Termination status : ", termination_status(model))
println("Primal status      : ", primal_status(model))
```

```julia
obj_value = objective_value(model)
p = value.(p)
x = M*p
```

```julia
Plots.plot(p)
Plots.title!("Control Force over Time (L1 Norm)")
```

```julia
# Plotting position over time (A)

Plots.plot(x)
Plots.title!("Position over Time (L1 Norm)")
```

## $l_2$ Solution

```julia
# Running least squares to solve for p

p = transpose(A)*inv(A*transpose(A))*y
# Calculating corresponding position vector

x = M*p

# Plotting control force vector over time

Plots.plot(p)
Plots.title!("Control Force over Time (L2 Norm)")
```

```julia
# Plotting position over time (A)

Plots.plot(x)
Plots.title!("Position over Time (L2 Norm)")
```

## $l_\infty$ Solution

```julia
model = Model(GLPK.Optimizer)

clctr = ones(10)

@variable(model, p[1:10])
@variable(model, t) # Helper variable

@objective(model, Min, t)

@constraint(model, sum(p) == 0)

@constraint(model, 9.5p[1]+8.5p[2]+7.5p[3]+6.5p[4]+5.5p[5]+4.5p[6]+3.5p[7]+2.5p[8]+1.5p[9]+0.5p[10] == 1)

@constraint(model, p .<= t)
@constraint(model, p .>= -t)

@constraint(model, t .>= 0)


print(model)
```

```julia
optimize!(model)
println("Termination status : ", termination_status(model))
println("Primal status      : ", primal_status(model))
```

```julia
obj_value = objective_value(model)
p = value.(p)
x = M*p
```

```julia
Plots.plot(p)
Plots.title!("Control Force over Time (L Infinity Norm)")
```

```julia
# Plotting position over time (A)

Plots.plot(x)
Plots.title!("Position over Time (L Infinity Norm)")
```
