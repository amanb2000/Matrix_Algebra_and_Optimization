```julia
using LinearAlgebra
using Plots
```

```julia
function f(n)
    return (-1/3)^(n-2)
end
```

```julia
function δ(n)
    if n == 0
        return 1
    end
    return 0
end
```

```julia
many = 10
```

```julia
ns = -4:-4+many
```

```julia
function g(m)
    sum = 0
    for n = 0:3
        sum += f(n)*δ(n-m)
    end
    return sum
end
```

```julia
gs = [];

for m = -5:5
    append!(gs, g(m));
end
    
```

```julia
plot(-5:5, gs)
```
