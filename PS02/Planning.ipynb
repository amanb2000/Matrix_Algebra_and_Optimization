```julia
# B: 3D Plots
    # From: https://www.youtube.com/watch?v=S47KNst7mS8&ab_channel=JohanJansson

x = y = -5:0.1:5

f(x, y) = (x-5)cos(y-5) - (y-5)sin(x-5)

# z(x, y) = x+sin(y)

Plots.surface(x, y, f, color = :gray, label="Surf 1", title="Hello")
# Plots.surface!(x, y, z, label="Surf2", color=:green)
```

```julia
# C: Contour Plots
    # From: http://docs.juliaplots.org/latest/generated/plotly/#plotly-ref22

x = 1:0.5:20
y = 1:0.5:10
f(x, y) = begin
        (3x + y ^ 2) * abs(sin(x) + cos(y))
    end

p1 = Plots.contour(x, y, f, fill = true)
p2 = Plots.contour(x, y, f)

Plots.plot(p1, p2)
```
